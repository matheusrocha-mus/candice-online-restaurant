```python
# Matheus Caetano Rocha - 09/28/2022


# Imported the "clear" function to clear terminal after user input.
import os
clear = lambda: os.system('cls')


# A function that only accepts "1" or "2" as valid input from the user.
# The user will stay on the function loop until he types a valid input.
def yes_no ():
    global inp
    print ("1) Yes \n 2) No \n")
    while True:
        inp = input ()
        if inp == "1" or inp == "2":
            break

        else:
            print("\n Please type 1 for 'Yes' or 2 for 'No'. \n")


# Same role as "yes_no", but with more accepted options for user input.
def choose_meal ():
    global inp
    while True:
        inp = input ()
        if inp == "0" or inp == "1" or inp == "2" or inp == "3" or inp == "4":
            break

        else:
            print("\n Please enter the number between 1 and 4 that corresponds to the desired option. If you want to cancel and exit, type 0. \n")


# A function called whenever the user chooses to exit the program.
def farewell ():
    print ("\n Thank you for your visit! \n")


# The indexes of theses arrays correspond to each meal option in the menu, with it's respective price and kcal value.
meal = ["Vegan", "Fish", "Chicken", "Beef", "Pineapple", "Diet icecream", "Diet mousse", "Chocolate mousse", "Tea", "Orange juice", "Melon juice", "Diet soda"]
price = [8.50, 12.30, 7.50, 10.50, 1.50, 3.50, 3.00, 2.50, 1.00, 2.50, 3.00, 2.00]
kcal = [180, 230, 250, 350, 75, 110, 170, 200, 20, 70, 100, 65]

print ("Hello! Welcome to the Candice Restaurant official website. Would you like to see the menu? \n \n")

yes_no ()
clear ()

if inp == "2":
    farewell ()

elif inp == "1":

    # The section below displays the menu.
    for I in range (12):
        if I == 0:
            print ("\n DISH \n ------------------------------- \n")

        elif I == 4:
            print ("\n DESSERT \n ------------------------------- \n")

        elif I == 8:
            print ("\n DRINK \n ------------------------------- \n")

        print (meal[I], " - $", format(price[I], ',.2f'), " - ", kcal [I], " kcal \n")

    print ("\n Would you like to make an order? \n \n")

    yes_no ()
    clear ()

    if inp == "2":
        farewell ()

    elif inp == "1":

        # This loop corresponds to the option given to the user at the end of his order to restart
        # the program as a "new customer". If he chooses so, the program returns to this point.
        while True:

            # The first time that the program runs, this code will create a file to store the user meal choices.
            # Meanwhile, each time that the user chooses to restart the program, this code will delete
            # all the content inside the file.
            file = open ("order.txt", "w")
            file.close()

            print ("\n Choose below which meal you would like to order: \n \n")
            print ("0) Cancel and exit \n \n 1) Dish \n 2) Dessert \n 3) Drink \n")
            
            while True:
                inp = input ()
                if inp == "0" or inp == "1" or inp == "2" or inp == "3":
                    break

                else:
                    print ("\n Please enter the number between 1 and 3 that corresponds to the desired option. If you want to cancel and exit, type 0. \n")

            # This is the meal choosing session. The user will only exit this section if he types "4"
            # to go to checkout or "0" to cancel and exit.
            while inp == "1" or inp == "2" or inp == "3":

                if inp == "1":
                    print ("\n Choose a dish: \n \n")
                    print ("0 - Cancel and exit \n \n")
                    for I in range(1, 5):
                        print (I, ") ", meal [I - 1], " - $", format(price[I - 1], ',.2f'), " - ", kcal [I - 1], " kcal \n")

                    choose_meal ()
                    file = open ("order.txt", "a")
                    temp = int (inp)
                    temp = temp - 1
                    mealChoice = str (temp)
                    file.write(mealChoice + "\n")
                    file.close()


                elif inp == "2":
                    print ("\n Choose a dessert:: \n \n")
                    print ("0 - Cancel and exit \n \n")
                    for I in range(1, 5):
                        print (I, ") ", meal [I + 3], " - $", format(price[I + 3], ',.2f'), " - ", kcal [I + 3], " kcal \n")

                    choose_meal ()
                    file = open ("order.txt", "a")
                    temp = int (inp)
                    temp = temp + 3
                    mealChoice = str (temp)
                    file.write(mealChoice + "\n")
                    file.close()


                elif inp == "3":
                    print ("\n Choose a drink: \n \n")
                    print ("0 - Cancel and exit \n \n")
                    for I in range(1, 5):
                        print (I, ") ", meal [I + 7], " - $", format(price[I + 7], ',.2f'), " - ", kcal [I + 7], " kcal \n")

                    choose_meal ()
                    file = open ("order.txt", "a")
                    temp = int (inp)
                    temp = temp + 7
                    mealChoice = str (temp)
                    file.write(mealChoice + "\n")
                    file.close()

                clear ()

                if inp == "0":
                    break

                elif inp != "0":

                    # With each user meal choice, the program stores the input on the "order.txt" file - which will then 
                    # be used next on the checkout session.
                    print ("\n Would you like to order anything else? \n \n")
                    print ("0) Cancel and exit \n \n 1) Dish \n 2) Dessert \n 3) Drink \n \n 4) Finish and go to checkout \n")
                    
                    while True:
                        inp = input ()
                        if inp == "0" or inp == "1" or inp == "2" or inp == "3" or inp == "4":
                            break

                        else:
                            print ("\n Please enter the number between 1 and 3 that corresponds to the desired option. If you want to cancel and exit, type 0. \n")

            clear ()

            if inp == "4":

                print ("\n CHECKOUT \n")
                print ("----------------------------------- \n")

                priceFinal = 0.0
                kcalFinal = 0

                # This is the checkout session. The user's inputs for meal choices are read one by one in the 
                # "order.txt" file, matched with the corresponding indexes on the "meal", "price" and "kcal"
                # arrays, added up and displayed to the user.
                file = open ("order.txt", "r")      
                
                # This loop ensures that all the meal choices will be read.
                while True:
                    line = file.readline ()
                    if line == "":
                        break
                    mealChoice = int (line)
                    print (meal [mealChoice], " - $", format(price[mealChoice], ',.2f'), " - ", kcal [mealChoice], " kcal \n")
                    priceFinal = priceFinal + price [mealChoice]
                    kcalFinal = kcalFinal + kcal [mealChoice]
                    
                file.close()

                print ("=================================== \n")

                print ("TOTAL: $", format(priceFinal, ',.2f'), " - ", kcalFinal, " kcal \n \n")

                print ("Would you like to make another order as a new customer? \n \n")

                yes_no ()
                clear ()


            # This corresponds to the "0) Exit" input option and the choice given to the user to restart
            # the program as a "new customer", as mentioned earlier.
            if inp == "2" or inp == "0": 
                farewell ()
                break

            else:
                priceFinal = 0.0
                kcalFinal = 0
```
